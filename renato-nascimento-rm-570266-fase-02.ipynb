## **Atributos Básicos**

5. **Horário Estimado de Chegada à Órbita** - Momento previsto para o módulo chegar à órbita de Marte.
6. **Criticidade da Carga** - Importância do módulo para a sobrevivência ou continuidade da missão.

8. **Risco de Pouso** - Chance de dificuldade durante o pouso: baixo, médio ou alto.
9. **Integridade Estrutural** - Percentual que representa condição física do módulo.
10. **Depêndencia Principa** - Indica se o módulo depende de outro para operar corretamente.
11. **Tempo Estimado de Pouso** - Duração aproximada da manobra de descida até a superfície.

# **2. Definir e representar regras de decisão usando portas lógicas**

## **Especificação de Condições Críticas para Autorização ou Bloqueio de Pouso do Módulo**

1. **C** - Combustível Suficiente - Quando o nível de combustível for maior ou igual a 75%.
2. **A** - Condições Atmosféricas Seguras - Quando não houver tempestade, vento extremo ou baixa visibildiade.
3. **L** - Área de Pouso Livre - Quando a zona de pouso estiver disponível e sem obstáculos.
4. **S** - Sensores Íntegros - QUando os sensores estiverem funcionando corretamente.
5. **E** - Energia Disponível - Quando o módulo possui energia suficiente para acionar sistemas de pouso.
6. **TEMP** - Temperatura Interna Controla - Quando a temperatura dos sistemas está dentro do limite seguro.
7. **PRES** - Pressão interna estável - Quando a pressão do módulo está dentro do padrão operacional.
8. **VEL** - Quando a velocidade está dentro do limite permitido.
9. **ANG** - Ângulo de entrada correto - Quando o ângulo de aproximação está adequado.
10. **EST** - Estabilidade Estrutural - Quando a estrutura do módulo está preservada.

```python
# Funções do algoritmo de busca 

def buscar_menor_combustivel(lista):
    if len(lista) == 0:
        return None 

    menor = lista[0]

    for modulo in lista: 
        if modulo["combustivel"] < menor ["combustivel"]:
            menor = modulo 

    return menor 

def buscar_maior_prioridade(lista): 
    if len(lista) == 0: 
        return None 

    # Quanto menor o número, maior será a prioridade 
    maior = lista[0]

    for modulo in lista: 
        if modulo["prioridade"] < maior["prioridade"]:
            maior = modulo 

    return maior 

def buscar_por_criticidade(lista, criticidade_desejada): 
    encontrados = [] 

    for modulo in lista:
        if modulo["criticidade"].lower() == criticidade_desejada.lower(): 
            encontrados.append(modulo) 

    return encontrados 
```

```python
# Funções de algoritmos de ordenaçãoabs
def ordenar_por_prioridade(lista): 
    # Selection Sort 
    copia = lista[:]

    for i in range(len(copia)): 
        indice_menor = i 

        for j in range(i + 1, len(copia)): 
            if copia[j]["prioridade"] < copia[indice_menor]["prioridade"]:
                indice_menor = j 

        copia[i], copia[indice_menor] = copia[indice_menor], copia[i] 

    return copia

def ordenar_por_combustivel(lista):
    #BubbleSort  para organizar o maior combustível para o menor 
    copia = lista[:]

    for i in range(len(copia)):
        for j in range(0, len(copia) - 1 - i):
            if copia[j]["combustivel"] < copia[j + 1]["combustivel"]: 
                copia[j], copia[j + 1] = copia[j + 1], copia[j] 

    return copia 
```

```python
# Utilizando o NOR da porta lógica 
# A porta NOR só retornara True quando as entradas forem False 
# Vai funcinar da seguinte forma:
# False = não há falha 
# True = há falha
def porta_nor(entradas):
    for entrada in entradas: 
        if entrada == True: 
            return False 

    return True 
```

```python
# Lógica do s Circuitos AND + NOR 
# Avaliando condições de pouso 
def avaliar_regras_logicas(modulo, atmosfera_segura, area_livre): 
    C = modulo["combustivel"] >= limites["combustivel_minimo"]
    A = atmosfera_segura == True 
    L = area_livre == True 
    S = modulo["sensores_integros"] == True 
    E = modulo["energia"] >= limites["energia_minima"] 
    TEMP = limites["temperatura_minima"] <= modulo["temperatura"] <= limites["temperatura_maxima"]
    PRES = limites["pressao_minima"] <= modulo["pressao"] <= limites["pressao_maxima"]
    VEL = modulo["velocidade"] <= limites["velocidade_maxima"]
    ANG = limites["angulo_minimo"] <= modulo["angulo"] <= limites["angulo_maximo"]
    EST = modulo["integridade"] >= limites["integridade_minima"]

    condicoes = {
        "C": C,
        "A": A,
        "L": L,
        "S": S,
        "E": E,
        "TEMP": TEMP,
        "PRES": PRES,
        "VEL": VEL,
        "ANG": ANG,
        "EST": EST
    }

    # Bloco AND de autorização de pouso - todas as condições precisam ser verdadeiras
    bloco_autorizacao_and = (
        C and
        A and
        L and 
        S and 
        E and 
        TEMP and 
        PRES and 
        VEL and 
        ANG and 
        EST
    )

    # Bloco de falhas - cada falha será ao contrário da condição esperada 
    falhas = {
        "Falhas_C": not C,
        "Falhas_A": not A,
        "Falhas_L": not L,
        "Falhas_S": not S,
        "Falhas_E": not E,
        "Falhas_TEMP": not TEMP, 
        "Falhas_PRES": not PRES,
        "Falhas_VEL": not VEL,
        "Falhas_ANG": not ANG,
        "Falhas_EST": not EST,
    }

    # Entradas da porta NOR
    entradas_bloqueio = []

    for falha in falhas.values():
        entradas_bloqueio.append(falha) 

    # Se nenhuma falha existir, NOR retorna True,
    # Se qualquer falhar existir, NOR retorna False
    bloco_bloqueio_nor = porta_nor(entradas_bloqueio)

    # O pouso só é autorizado se AND = TRUE e NOR = TRUE
    saida_final = bloco_autorizacao_and and bloco_bloqueio_nor 

    falhas_detectadas = []

    for nome_falha, valor in falhas.items(): 
        if valor == True:
            falhas_detectadas.append(nome_falha)

    return saida_final, condicoes, falhas_detectadas, bloco_autorizacao_and, bloco_bloqueio_nor
```

```python
# Simulando a fila de pouso

def simular_pouso(fila_pouso, horario_atual, atmosfera_segura, area_livre): 
    #Listas Auxiliares
    modulos_pousados = []
    modulos_em_espera = []
    modulos_em_alerta = [] 

    # Pilha de revisão
    pilha_revisao = []

    print("\n===== SIMULAÇÃO DE POUSO DA NAVE AURORA =====")

    while len(fila_pouso) > 0: 
        modulo = fila_pouso.pop(0) # A fila usa pop(0), para retirar o primeiro elemento
        print("\nMódulo analisado:", modulo["nome"])

        # Verificando se o módulo chegou à órbita
        if modulo["chegada_orbita"] > horario_atual:
            print("Resultado: EM ESPERA")
            print("Motivo: ainda não chegou à órbita no horário atual.")
        else:
            autorizado, condicoes, falhas, bloco_and, bloco_nor = avaliar_regras_logicas(
                modulo,
                atmosfera_segura,
                area_livre
            )

            # If do pouso autorizado
            if autorizado:
                modulos_pousados.append(modulo) 
                print("Resultado: POUSO AUTORIZADO")
                print("Bloco AND:", int(bloco_and), "| Bloco NOR:", int(bloco_nor))

            #ELIF do pouso bloqueado ou módulo em alerta
            elif len(falhas) > 0 or modulo["status"] == "Atenção": 
                modulos_em_alerta.append(modulo)
                pilha_revisao.append(modulo)
                print("Resultado: POUSO ADIADO/BLOQUEADO")
                print("Bloco AND:", int(bloco_and), "| Bloco NOR:", int(bloco_nor))
                print("Falhas detectadas:", ", ".join(falhas))

            #ELSE caso não se encaixe nos casos anteriores
            else:
                modulos_em_espera.append(modulo)
                print("Resultado: EM ESPERA")
                print("Motivo: aguardando nova verificação.")

        return modulos_pousados, modulos_em_espera, modulos_em_alerta, pilha_revisao
        
```

```python
# Função auxiliar para mostrar os resultados
def mostrar_nomes(titulo, lista):
    print("\n", titulo)

    if len(lista) == 0:
        print("Nenhum módulo.")
    else: 
        for modulo in lista:
            print("-", modulo["nome"])
```

```python
modulos = criar_modulos()

# Lista principal com todos os módulos cadastrados 
lista_modulos = modulos 

# Fila principal organizada por prioridade 
fila_principal = ordenar_por_prioridade(lista_modulos)

# Busca por características
menor_combustivel = buscar_menor_combustivel(lista_modulos)
maior_prioridade = buscar_maior_prioridade(lista_modulos)
modulos_criticos = buscar_por_criticidade(lista_modulos, "Crítica")

# Ordenação alternativa por combustível
fila_por_combustivel = ordenar_por_combustivel(lista_modulos)

print("Módulo com menor combustível:", menor_combustivel["nome"], "-", str(menor_combustivel["combustivel"]) + "%")
print("Módulo com maior prioridade:", maior_prioridade["nome"], "- prioridade", maior_prioridade["prioridade"])

print("\nMódulos com criticidade Crítica:")
for modulo in modulos_criticos:
    print("-", modulo["nome"])

print("\nFila principal organiada por prioridade: ")
for modulo in fila_principal:
    print(modulo["prioridade"], "-", modulo["nome"])

print("\nFila reorganizada por Combustível:")
for modulo in fila_por_combustivel: 
    print(modulo["combustivel"], "% -", modulo["nome"])

# Condições simuladas do ambiente externo 
horario_atual = "09:30"
atmosfera_segura = True 
area_livre = True 

# Simulação de Pouso
pousados, espera, alerta, revisao, = simular_pouso(
    fila_principal,
    horario_atual,
    atmosfera_segura,
    area_livre
)

# Resultados finais
mostrar_nomes("MÓDULOS POUSADOS:", pousados) 
mostrar_nomes("MÓDULOS EM ESPERA:", espera)
mostrar_nomes("MÓDULOS EM ALERTA:", alerta)

print("\nPILHA DE REVISÃO:")

if len(revisao) > 0:
    print("Topo da pilha:", revisao[-1]["nome"])
else:
    print("Nenhum módulo na pilha.")
```

# **4. Modelar funções matemáticas aplicadas ao pouso ou estabilização da base**

## **Função Quadrática**

Durante a descida até a superfície de Marte, podemos relatar esse comportamento representado pela função quadrática, pois a altura muda com o tempo sob influência da velocidade inicial e da aceleração gravitacional.

h(t) = h0 − v0t − 1/2 gt2

Explicação da função:
- h(t) - altura do módulo no instante t
- h0 - altura inicial do módulo
- v0 - velocidade inicial de descida
- g - aceleração gravitacional simulada
- t - tempo de descida

Quanto o t aumenta, h(t) diminui. Quando h(t) se aproxima de zero, significa que o módulo está próximo da superfície.

Essa função ajudará o sistema a decidir o momento correto de acionar os retrofoguetes. Se a altura estiver diminuindo muito rápido, o MGPEB pode bloquear o pouso ou antecipar a frenagem.

```python
import numpy as np
import matplotlib.pyplot as plt

# Altura do Módulo em Função do Tempo de Descida 

h0 = 10000 # altura inicial em metros
v0 = 50    # velocidade incial de descida em m/s
g = 3.71   # gravidade de Marte em m/s2

# tempo de 0 a 60 segundos
t = np.linspace(0, 60, 200)

# Função da altura
h = h0 - v0*t - 0.5*g*(t ** 2)

# Plotando
plt.figure(figsize=(8, 5))
plt.plot(t, h) 
plt.title("Altura do Módulo em Função do Tempo - Função Quadrática")
plt.xlabel("Tempo de descida (s)")
plt.ylabel("Altura (m)")
plt.grid(True)
plt.show()
```

## **Função Afim**

Durante a descida, podemos entender através da função afim, que quanto maior a velocidade de descida, maior tende a ser o esforço dos sistemas de frenagem e propulsão.

C(v) = av + b

Explicação da função: 
- C(v) - consumo de combustível
- v - velocidade do módulo
- a - taxa de aumento do consumo conforme a velocidade cresce
- b - consumo mínimo dos sistemas ligados

Essa função gera uma reta crescente. Quand oa velocidade aumenta, o consumo também aumenta.
Se a velocidade diminui, o consumo também tende a diminuir. Se a velocidade aumenta demais, o consumo pode ultrapassar o limite seguro.

O MGPEB pode usar essa previsão para verificar se o módulo tem combustível suficiente para completar a descida. Dessa forma, o pouso não depende apenas do combustível atual, mas também do combustível necessário para finalizar a manobra.

```python
# Consumo de Combustível em Função da Velocidade 

a = 0.05 # taxa de consumo
b = 10   # consumo mínimo

# Velocidade de 0 a 500
v = np.linspace(0, 500, 200)

# Função do consumo 
c = a*v + b 

# Plotando
plt.figure(figsize=(8, 5))
plt.plot(v, c)
plt.title("Consumo de Combustível em Função da Velocidade - Função Afim")
plt.xlabel("Velocidade (m/s)")
plt.ylabel("Consumo Estimado")
plt.grid(True)
plt.show()
```

## **Função Trigonométrica**

Para a estabilização da base, a energia é fundamental. O módulo de Energia pode usar painéis solares auxiliares, e a geração varia conforme o horário do dia. Podemos representar este fenômeno pela função trigonométrica, pois a incidência solar cresce e diminui ao longo do dia.

E(t) = Emax * sin(πt/D)

Explicação da função:
- E(t) - energia gerada no instant t
- Emax - geração máxima de energia
- t - tempo desde o nascer do Sol
- D - duração do período iluminado do dia

Essa função começa próximo do zero no início do dia e cresce até atingir um valor máximo no meio do período iluminado e depois diminui novamente.

Quando o Sol está baixo, a geração é pequena. Quando o Sol está mais alto, a geração aumenta. No fim do dia, a energia volta diminuir.

Essa função ajuda o sistema a limitar a quantidade de módulos pousando ao mesmo tempo.
Se a energia disponível estiver baixa, o MGPEB pode permitir apenas um pouso por vez. Se a geração estiver alta, pode liberar mais operações.

```python
# Geração de Enenergia ao Longo do Dia - Função Trigonométrica 

Emax = 100 # geração máxima
D = 24     # duração do período analisado (horas)

# Tempo de 0 a 24 horas
t_energia = np.linspace(0, 24, 200)

# Função de geração de energia 
energia = Emax * np.sin(np.pi * t_energia / D) 

# Energia não pode ser negativa, portanto, faremos a correção
energia = np.maximum(energia, 0)

# Plotando

plt.figure(figsize=(8, 5))
plt.plot(t_energia, energia)
plt.title("Geração de Energia ao Longo do Dia - Função Trigonométrica")
plt.xlabel("Tempo (horas)!")
plt.ylabel("Energia Gerada")
plt.grid(True) 
plt.show()
```

## **Função Exponencial de Estabilização**

Após o pouso, os módulos precisam estabilizar a temperatura, pressão e sistemas internos. A temperatura pode se aproximar gradualmente de um valor seguro. Esse comportamento será representado pela função exponencial de estabilização, mesmo não estando dentro do conteúdo, me pareceu a mais paupável para esse fenômeno.

T(t) = Ts + (T0 − Ts) e − kt
Explicação da função:
- T(t) - temperatura no instant t
- Ts - temperatura segura desejada
- T0 - temperatura inicial
- k - taxa de estabilização térmica
- t - tempo

Essa função mostra a temperatura se aproximando aos poucos de um valor seguro. No início, a mudança pode ser mais rápida; depois, vai ficando mais lenta até estabilizar.

Se a temperatura inicial estiver muito alta, ela diminui até se aproximar da temperatura segura. Se estiver muito baixa, pode aumentar até atingir a faixa ideal.

Se a temperatura ainda não estiver estabilizada, o MGPEB pode colcoar o módulo em espera antes de permitir conexão com outros módulos da base.

```python
# Temperatura Interna Durante a Estabilziação - Função Exponencial de Estabilização 

T0 = 80  # temperatura inicial
Ts = 22  # temperatura segura desejada
k = 0.15 # taxa de estabilização

# Tempo de 0 a 30 minutos 
t_temp = np.linspace(0, 30, 200)

# Função da temperatura 
temperatura = Ts + (T0 - Ts) * np.exp(-k * t_temp)

# Plotando
plt.figure(figsize=(8, 5))
plt.plot(t_temp, temperatura)
plt.title("Temperatura Interna Durante a Estabilização - Função Exponencial de Estabilização")
plt.xlabel("Tempo (min)")
plt.ylabel("Temperatura (°C)")
plt.grid(True) 
plt.show()
```

## **História e Evolução**

Os primeiros computadores de propósito geral surgiram em um contexto de grande demanda por cálculos científicos, militares e industriais. Máquinas como o ENIAC, desenvolvido na década de 1940, demonstraram que circuitos eletrônicos poderiam ser utilizados para executar operações matemáticas em alta velocidade. Esses computares eram enormes, consumiam grande quantidade de energia, exigiam manuteção constante e possuíam limitações severas de memória e programação. Ainda assim, eles estabeleceram um marco importante: a possibilidade de construir máquinas capazes de processar informações a partir de instruções lógicas e aritméticas.

A passagem dos computadores de sala para os sistemas embarcados foi possível graças à miniaturização dos componentes eletrônicos. Um marco essencial foi a inveção do transistor em 1947, no Bell Labs, que susbstituiu progressivamente as válvulas eletrônicas por dispositivos menores, mais eficientes e mais confiáveis. O transistor permitiu reduzir tamanho, calor e consumo, viabilizando circuitos mais complexos. Depois vieram os circuitos integrados e, em seguida, os microprocessadores. A Intel descreve o 4004, lançado em 1971, como um avanço conceitual decisivo: um microchip programável capaz de concentrar lógica computacional em um dispositivo extremamente compacto.

Outro passo importante foi a padronização arquitetural de computadores. O IBM System/360, lançado em 1964, foi projetado como uma família de computadores compatíveis entre si, destinada a diferentes tipos de usuários e aplicações, consolidando a ideia de uma arquitetura comum para múltiplos usos. Essa evolução tornou possível pensar o computador não apenas como uma máquina isolada, mas como uma arquitetura adaptável e diferentes contextos.

No campo espacial, essa adaptação aparece de forma clara no Apollo Guidance Computer, projetado para fornecer os cálculos necessáriso de orientação, navegação e controle dos módulos da missão Apollo. O Apollo Guidance Computer é um exemplo histórico de sistema embarcado: não era um computador destinado a tarefas gerais de escritório, mas um equipamento integrado a uma missão específica, com reponsabilidade direta sobre operações críticas.

O MGPEB se aproxima desse conceito de sistema embarcado. Diferentemente de um computador de uso geral, que pod executar uma grande variedade de tarefas, o MGPEB possui uma finalidade bem definida: gerenciar a autorização ou bloqueio do pouso dos módulos da nave Aurora. Para isso, ele utiliza regras lógicas, estruturas de dados lineares, algoritmos de busca e ordenação, além de funções matemáticas aplicadas ao pouso e à estabilização da base. Essa especialização é típica dos sistemas embarcos, nos quais o objetivo principal não é a generalidade, ma sim a confiabilidade, e previsibilidade e a eficiência.

## **Sistemas Embarcados de Alta Confiabilidade**

Sistemas espaciais precisam operar em ambientes extremos, com pouca ou nenhuma possibilidade de manutenção física. Por isso, a computação embarcada utilizada em missões espaciais constuma priorizar confiabilidade, redundância e previsibilidade. A NASA informa que o rover Perseverance utiliza computadores RAD750 resistentes à radiação, com dois módulos principais de computação, de modo que um possa atuar como reserva do outro. O MGPEB, deve ser pensado dentro dessa filosofia. A equipe não deve escolher algoritmos apenas por sofisticação, mas por clareza, baixo custo e facilidade de verificação. Um sistema de pouso não pode depender de uma lógica obscura ou difícil de validar.

Em uma missão para Marte, as limitações de hardware são decisivas para o projeto computacionao. A primeira limitação relevante é a memória. Sistemas espaciais não podem depende de armazenamento ilimitado, pois a memória embarcada precisa ser resistente, confiável e, em muitos casos, protegida contra falhas causadas por radiação. No caso do MGPEB, a lógica booleana baseada em condições como combustível suficiente, energia disponível, temperatura controlada, pressão estável e estabildiade estrutural permite decisões rápidas e energeticamente eficientes.

Por esse motivo, sistemas embarcados em missões espaciais precisam ser projetados com estratégias de segurança, redundância e verificação. No MGPEB, essa preocupação aparece na inclusão de condições críticas como sensores íntegros e estabilidade estrutural. A simplicidade do modelo não representa limitação conceitual, mas uma esocolha compatível com ambientes críticos, nos quais previsibilidade, economia de recursos e segurança são prioridades fundamentais.

# **6. Incorporar princípios ESG à concepção da base Aurora**

## **Introdução**

A implantação da base Aurora em Marte não deve ser compreendida apenas como um desafio tecnológico, mas também como um projeto de responsabilidade ambiental, social e governança. Mesmo em um ambiente extraterrestre, onde ainda não há sociedade humana estabelicida nem ecossistemas conhecidos da mesma forma que na Terra, as decisões da missão precisam seguir critérios éticos, sustentáveis e transparentes. Nesse sentido, os princípios ESG servem como orientação para que a colônia opere com responsabilidade, evitando que a exploração espacial repita erros históricos cometidos em processos de ocupação, extração e desenvolvimento tecnológico na Terra.

## **Reflexão**

A escolha da área de pouso da base Aurora deve considerar, antes de tudo, a segurança da missão e a preservação do ambiente marciano. O local escolhido precisa apresentar estabilidade do solo, baixa inclinação, ausência de obstáculos, menor risco de tempestades de poeira e condições adequadas para comunicação e geração de energia. No entanto, a decisão não pode ser baseada apenas na conveniência operacional. Também é necessário avaliar o possível impacto da presença humana e dos módulos de pouso sobre o ambiente marciano.

Mesmo que Marte não possua vida complexa conhecida, a missão deve adotar o princípio da precaução. Isso significa evitar áreas de grande interesse científico, como regiões com possível presença de gelo subterrâneo, formações geológicas raras ou locais que possam preservar evidências de vida passada. A instalação da base em uma região inadequada poderia contaminar amostras, alterar o solo ou prejudicar futuras pesquisas científicas.

Outra questão que deve ser levado em conta é a energia, um dos elementos mais críticos para a sobrevivência da base Aurora. Sem energia, os módulos de habitação, controle ambiental, comunicação, suporte médico, armazenamento de dados e laboratório científico não conseguem operar. Por isso, a gestão energética precisa ser pensada de forma sustentável e eficiente.

A base poderia utilziar uma combinação de fontes, como painéis solares, baterias de armazenamento e, em caso de necessidade, sistemas nucleares compactos. A governança ambiental exige que o uso de energia seja planejado por prioridade. Sistemas vitais, como suporte de vida, controle térmico, comunicação e segurança, devem ter preferência sobre sistemas secundários.

Outro critério extremamente importante para a sustentabilidade da base Aurora depende da capacidade de reduzir desperdicíos e reaproveitar materiais. Em Marte, o envio constante de suprimentos da Terra seria caro, demorado e arriscado. Por isso, a colônia precisa operar com o máximo possível de reaproveitamento.

A gestão de resíduos deve considerar separação, reciclagem, reaproveitamento e armazenamento seguro. Resíduos orgânicos poderiam ser tratados para produção de fertilizantes em sistemas agrícolas controlados. Água utilizada em higiene, alimentação e processos laboratoriais deveria ser filtrada e reutilizada. Peças metálicas, plásticas e componentes eletrônicos poderiam ser reaproveitados em manutenção e fabricação local.

Uma base sustentável deveo perar com lógica de ciclo fehcado, isto é, buscando transformar resíduos em novos recursos. A água, por exemplo, não deve ser vista como recurso descartável, mas como parte de um sistema contínuo de uso, filtragem, reaproveitamento e monitoramento.

Outro ponto é o uso de recursos disponíveis em Marte, conhecido como utilização de recursos locais. A base Aurora poderia estudar formas de usar gelo marciano para produção de água, oxigênio e combustível, desde que fosse feito de maneira responsável tanto ambiental como científica.

A dimensão social do ESG envolve a forma como a colônioa protege as pessoas que vivem e trabalham nela. Em uma base marciana, a tripulação estaria submetida a isolamento, radiação, confinamento, risco operacional, atraaso de comunicação com a Terra e dependência extrema da tecnologia.

Por isso, a base Aurora deve priorizar a saúde física e mental, em condições dignas de convivência, divisão justa de tarefas, acesso a suporte médico e mecanismos de participação nas decisões internas.

O módulo de suporte médico teria papel central, pois permitiria monitoramento a saúde, telemedicina, exames básicos e resposta a emergências. Já o módulo de habitação deveria garantir conforto mínimo, controle ambiental, segurança térmica, privacidade e espaços de convivência.

A governana organiza como as decisão são tomadas, registradas, auditadas e justificadas. Em uma base como a Aurora, decisões sobre pouso, energia, uso de recursos, segurança, saúde e pesquisa científica não podem ser tomadas de forma improvisada.

O MGPEB pode contribuir diretamente para a governança porque registra critérios objetivos para autorização ou bloqueio de pouso. Suas regras booleanas tornam a decisão mais transparente: o pouso só é autorizado se todas as condições críticas forem verdadeiras e nenhuma falha for detectada no bloco de bloqueio.

## **Conclusão**

A incorporação de princípios ESG à base Aurora demonstra que a exploração espacial deve ser conduzida com responsabilidade ambiental, social e de governança. A escolha da área de pouso deve considerar não apenas a segurança técnica, mas também a preservação científica e ambiental do território marciano. A gestão de energia, resíduos e recursos locais deve seguir princípios de eficiência, reaproveitamento e redução de desperdícios. A dimensão social exige atenção à saúde, segurança, bem-estar e participação da tripulação.

Do ponto de vista da governança, o MGPEB representa uma ferramenta importante, pois transforma decisões críticas em regras claras, verificáveis e documentáveis. Ao usar lógica booleana, estruturas de dados e critérios objetivos, o sistema contribui para uma operação mais transparente e segura.

Assim a base Aurora não deve ser projetada paenas para funcionar em Marte, mas para funcionar de maneira ética, sustentável e responsável. Mesmo em um ambiente de fronteira, a tecnologia precisa estar subordinada a princípios de cuidado, transparência e responsabilidade coletiva.

# **ENTREGÁVEIS - Descrição das Estruturas de Dados - Listas, Filas e Pilhas**

## **LISTAS**

As listas foram utilizadas como estrutura principal para armazenar os módulos da nave Aurora. Cada módulo foi representado por um dicionário, contendo informações como nome, prioridade, combustível, massa, criticidade, status, integridade, temperatura, pressão, velocidade e ângulo.

Exemplo retirado do código: 
- modulos = criar_modulos()
- lista_modulos = modulos

Nesse trecho, a função criar_modulos() retorna uma lista contendo todos os módulos cadastrados. Essa lista é armazenada na variável "modulos" e depois atribuída à variável "lista_modulos".

A lista funciona como uma coleção geral dos módulos da missão.
Um exemplo simplificado de um módulo para base, dentro da lista: 
-         {
            "id": 13,
            "nome": "Lógistica",
            "prioridade": 11,
            "combustivel": 72,
            "massa": 13500,
            "criatividade_carga": 3,
            "criticidade": "Média",
            "chegada_orbita": "09:40",
            "status": "Atenção",
            "risco_pouso": "Médio",
            "integridade": 93,
            "dependecia": "Habitação",
            "tempo_pouso": 23,
            "sensores_integros": True,
            "energia": 76,
            "temperatura": 24,
            "pressao": 100,
            "velocidade": 440,
            "angulo": 12
        },

Isso significa que cada posição da lista guarda um módulo completo com seus atributos.

**Lista que armazena os módulos**

A lista principal guarda todos os módulos, independentemente do estado deles.
- lista_modulos = modulos

Essa lista é usada em várias partes do projeto, por exemplo: 

- menor_combustivel = buscar_menor_combustivel(lista_modulos)
- maior_prioridade = buscar_maior_prioridade(lista_modulos)
- modulos_criticos = buscar_por_criticidade(lista_modulos, "Crítica")

Nesse caso, a lista é usada como base para os algoritmos de busca. 

**Lista de módulos pousados**

Durante a simulação, foi criada uma lista para armazenar os módulos que conseguram pousar com sucesso.

- modulos_pousados = []

Essa lista começa vazia porque, no início da simulação, nenhum módulo pousou ainda. Quando um módulo é aprovado pelas regras lógicas do MGPEB, ele é adicionado à lista de pousados.

- if autorizado:
    - modulos_pousados.append(modulo)
    - print("Resultado: POUSO AUTORIZADO")

Aqui o método .append() adicionad o módulo ao final da lista. 

- modulos_pousados.append(modulo)

Se o módulo analisado for o Módulo de Segurança, ele passa a fazer parte da lista de módulos pousados. Essa lista serve como um histórico dos módulos que concluíram o pouso.

**Lista de Módulos em Espera**

Também foi criada uma lista para armazenar os módulos que ainda não podem pousar, mas que não necessariamente apresentam falha grave.

- modulos_em_espera = []

Um módulo pode entrar nessa lista quando ainda não chegou à órbita no horário atual: 

- if modulo["chegada_orbita"] > horario_atual:
    - modulos_em_espera.append(modulo)
    - print("Resultado: EM ESPERA")
    - print("Motivo: ainda não chegou à órbita no horário atual.")

Nesse caso, o MGPEB não bloqueia definitivamente o pouso. Ele apenas coloca o módulo em espera: 

- Horário atual: 09:30
- Módulo Laboratório Científico chega à órbita: 09:50

Como o módulo ainda não chegou à órbita, ele não pode ser analisado para pouso. Então ele é enviado para a lista "modulos_em_espera". Essa estrutura é importante porque representa módulos que poderão ser avaliados novamente depois.

**Lista de Módulos em ALerta**

A lista de alerta armazena os módulos que apresentaram alguma falha ou condição insegura.

- modulos_em_alerta = []

Quando o sistema detecta falhas nas condições booleanas, o módulos é enviado para essa lista:

- elif len(falhas) > 0 or modulo["status"] == "Atenção":
    - modulos_em_alerta.append(modulo)
    - pilha_revisao.append(modulo)
    - print("Resultado: POUSO ADIADO/BLOQUEADO")

Essa lista tem uma função importante: separar os módulos problemáticos dos módulos que pousaram ou que apenas aguardam horário:

- Módulo: Estação de Cozinha
- Combustível: 68%
- Condição C: False
- Resultado: POUSO ADIADO/BLOQUEADO

Como o combustível está abaixo de 75%, esse módulo entra em alerta. 

## **FILAS**

A fila foi utilizada para representar a ordem de pouso dos módulos. A lógica da fila segue o princípio: FIFO - First In, First Out. Ou seja, o primeiro módulo da fila é o primeiro a ser analisado.

No projeto, a fila principal foi criada a partir da lista de módulos ordenada por prioridade: 

- fila_principal = ordenar_por_prioridade(lista_modulos)

A função "ordenar_por_prioridade()" organiza os módulos do menor número de prioridade para o maior. No projeto, quanto menor o número, maior a prioridade: 

- Prioridade 1 → Segurança
- Prioridade 2 → Navegação
- Prioridade 3 → Comunicação
- Prioridade 4 → Energia

Depois da ordenação, a fila representa a sequência em que o MGPEB analisará os módulos para pouso.

**Retirada de Elementos da FIla**

Dentro da função simular_pouso(), o sistema retira sempre o primeiro módulo da fila:

- while len(fila_pouso) > 0:
    - modulo = fila_pouso.pop(0)

Desta forma "modulo = fila_pouso.pop(0)" remove o primeiro elemento da lista, simulando o comportamento de uma fila.

Por exemplo, se a fila estiver assim:
- [Segurança, Navegação, Comunicação, Energia]

Ao sexecutar "modulo = fila_pouso.pop(0)" o módulo retirado será "Segurança" e a fila passará a ficar assim: 
- [Navegação, Comunicação, Energia]

Isso representa a lógica de uma fila de pouso: o primeiro módulos aguardando autorização é o primeiro a ser analisado.

## **PILHAS**

A pilha foi utilizada para armazenar módulos que precisam passar por revisão. A lógica da pilha segue o princípio LIFO - Last in, First Out. Ou seja, o último módulo inserido na pilha será o primeiro a ser retirado. No projeto a pilha foi criada assim: "pilha_revisao = []". Ela começa vazia e recebe módulos quando o pouso é bloqueado ou quando alguma falha é detectada.

**Inserção na Pilha**

Quando um módulo apresenta falha, ele é enviado para a pilha de revisão: 

- elif len(falhas) > 0 or modulo["status"] == "Atenção":
    - modulos_em_alerta.append(modulo)
    - pilha_revisao.append(modulo)
    - print("Resultado: POUSO ADIADO/BLOQUEADO")

A linha principal neste caso é "pilha_revisao.append(modulo)". O método .append() adiciona o módulo no topo da pilha. Por exemplo: 

- Pilha vazia:
- []

- Módulo Logística apresenta falha:
- [Logística]

- Depois Estação de Cozinha apresenta falha:
- [Logística, Estação de Cozinha]

Nesse caso, o topo da pilha é "Estação de Cozinha", por que foi o último módulo inserido.

**Consultando o Topo da Pilha**

No final do código o sistema mostra o topo da pilha de revisão: 

- if len(revisao) > 0:
    - print("Topo da pilha:", revisao[-1]["nome"])
- else:
    - print("Nenhum módulo na pilha.")

A linha "revisao[-1]" acessa o último elemento da lista, que representa o topo da pilha. Se a pilha estiver asssim:

- [Logística, Estação de Cozinha]

Então "revisao[-1]["nome"]" retorna "Estação de Cozinha". Isso significa que o último módulo enviado para revisão é o primeiro que deve receber atenção da equipe.
